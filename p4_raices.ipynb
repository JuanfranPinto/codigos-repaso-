```python
from scipy import optimize


def f(x):
    return x**5 + 5.25 * x**4 + 4.125 * x**3 - 9.125 * x**2 - 14.625 * x - 5.625

rangos = [(-4, 5), (-9, 3), (-2, 14)]

for r in rangos:
    raiz = optimize.bisect(f, r[0], r[1])
    print(f"Raíz encontrada en el rango {r}: {raiz}")
```

```python
## este no es borrar 
from scipy import optimize
import numpy as np

def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

# Paso 1: búsqueda de intervalos con cambio de signo
xs = np.arange(-6, 4, 0.5)
rangos = []

for i in range(len(xs) - 1):
    if f(xs[i]) * f(xs[i+1]) < 0:
        rangos.append((xs[i], xs[i+1]))

# Paso 2: aplicar bisección en cada rango
for r in rangos:
    raiz = optimize.bisect(f, r[0], r[1])
    print(f"Raíz encontrada en el rango {r}: {raiz}")
```

```python
def f(x):
    return x**5 + 5.25 * x**4 + 4.125 * x**3 - 9.125 * x**2 - 14.625 * x - 5.625


def fprime(x):
    return 5 * x**4 + 21 * x**3 + 12.375 * x**2 - 18.25 * x - 14.625


rangos_newton = [(-4, 5), (-9, 3), (-2, 14)]
for r in rangos_newton:
    x0 = (r[0] + r[1]) / 2
    raiz = optimize.newton(f, x0=x0, fprime=fprime)
    print(f"Raíz encontrada en el rango {r} usando el método de Newton: {raiz}")
```

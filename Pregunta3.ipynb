Usando los métodos estudiados, determine todas las raíces de la ecuación:

$x^5+5.5x^4+4.5x^3-9.5x^2-15.5x-6 = 0$

```python
import numpy as np
from scipy import optimize

# Definición del polinomio
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

# Derivada del polinomio
def df(x):
    return 5*x**4 + 21*x**3 + 12.375*x**2 - 18.25*x - 14.625

# Función para encontrar raíces reales usando Newton
def encontrar_raices():
    raices = []
    
    # Diferentes valores iniciales
    puntos_iniciales = np.linspace(-10, 10, 40)
    
    for x0 in puntos_iniciales:
        try:
            raiz = optimize.newton(f, x0, fprime=df, tol=1e-10, maxiter=100)
            
            # Verificar que sea real
            if np.isreal(raiz):
                raiz = float(np.real(raiz))
                
                # Evitar repetir raíces
                if all(abs(raiz - r) > 1e-5 for r in raices):
                    raices.append(raiz)
        
        except:
            pass
    
    # Orden ascendente
    raices.sort()
    
    # Completar hasta 6 valores con 0
    while len(raices) < 6:
        raices.append(0)
    
    return raices[:4]

resultado = encontrar_raices()

for i, r in enumerate(resultado, start=1):
    print(f"x_{i} = {r}")
```

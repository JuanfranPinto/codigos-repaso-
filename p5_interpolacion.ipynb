```python
from src import ajustar_min_cuadrados
def der_parcial_4(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = c_3 = c_2 = c_1 = c_0 = c_ind = 0
    for x, y in zip(xs, ys):
        c_4 += x**8 
        c_3 += x**7 
        c_2 += x**6 
        c_1 += x**5 
        c_0 += x**4 
        c_ind += x**4 * y
    return (c_4, c_3, c_2, c_1, c_0, c_ind)

def der_parcial_3(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = c_3 = c_2 = c_1 = c_0 = c_ind = 0
    for x, y in zip(xs, ys):
        c_4 = x**7 
        c_3 += x**6 
        c_2 += x**5 
        c_1 += x**4 
        c_0 += x**3 
        c_ind += x**3 * y
    return (c_4, c_3, c_2, c_1, c_0, c_ind)

def der_parcial_2(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = c_3 = c_2 = c_1 = c_0 = c_ind = 0
    for x, y in zip(xs, ys):
        c_4 = x**6 
        c_3 += x**5 
        c_2 += x**4 
        c_1 += x**3 
        c_0 += x**2 
        c_ind += x**2 * y
    return (c_4, c_3, c_2, c_1, c_0, c_ind)

def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = c_3 = c_2 = c_1 = c_0 = c_ind = 0
    for x, y in zip(xs, ys):
        c_4 = x**5 * y
        c_3 += x**4 * y
        c_2 += x**3 * y
        c_1 += x**2 * y
        c_0 += x 
        c_ind += x * y
    return (c_4, c_3, c_2, c_1, c_0, c_ind)

def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float, float, float, float]:
    c_4 = c_3 = c_2 = c_1 = c_0 = c_ind = 0
    for x, y in zip(xs, ys):
        c_4 = x**4 
        c_3 += x**3 
        c_2 += x**2 
        c_1 += x 
        c_0 += 1
        c_ind += y
    return (c_4, c_3, c_2, c_1, c_0, c_ind)

# UTILICE:
# ajustar_min_cuadrados()
a = ajustar_min_cuadrados(xs, ys, gradiente = [der_parcial_0, der_parcial_1, der_parcial_2, der_parcial_3, der_parcial_4])

def ajuste(x):
    return a[0]*x**4 + a[1]*x**3 + a[2]*x**2 + a[3]*x + a[4]

error_rel = 0
for i in range(len(xs)):
    error_rel += abs((ys[i] - ajuste(xs[i]))/ys[i])
error_rel = error_rel / len(xs)
print("error relativo promedio:", error_rel)
```
